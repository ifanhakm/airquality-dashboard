## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
- Stasiun mana yang paling tinggi suhunya?
```

```python
temp_by_station = combined_df.groupby(by="station").agg({"TEMP": "mean"}).reset_index()

temp_by_station_sorted = temp_by_station.sort_values(by="TEMP", ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(
    y="station", 
    x="TEMP",
    data=temp_by_station_sorted
)
plt.title("Stasiun Dengan Rata-Rata Suhu Tertinggi", loc="center", fontsize=15)
plt.xlabel('Temperature (°C)')
plt.ylabel('Station')
plt.show()
```

### Pertanyaan 2:

```python
- Bagaimana kenaikan gas Karbon dalam beberapa tahun tersebut?
```

```python
combined_df['year'] = combined_df['year'].astype(int)
```

```python
co_per_year = combined_df.groupby('year').agg({'CO': 'mean'}).reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(
    x='year',
    y='CO',
    data=co_per_year,
    marker='o',
    color='b'
)
plt.title('Kenaikan Gas Karbon (CO) per Tahun', fontsize=16)
plt.xlabel('Tahun')
plt.ylabel('Konsentrasi CO (mg/m³)')
plt.grid(True)
plt.show()
```

## Conclusion

- Stasiun dengan suhu rata-rata tertinggi dipegang oleh Stasiun Gucheng.
- Kenaikan gas Karbon(CO) mengalami penurunan di tahun 2015 ke 2016, namun melonjak tinggi di tahun 2017.
